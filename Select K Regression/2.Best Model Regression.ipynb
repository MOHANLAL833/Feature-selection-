```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
import pickle
```

```python
def selectkbest(indep_X,dep_Y,n):
        test = SelectKBest(score_func=chi2, k=n)
        fit1= test.fit(indep_X,dep_Y)
        selectk_features = fit1.transform(indep_X)
        selectk_features = fit1.transform(indep_X)
        selected_indices = fit1.get_support(indices=True) 
        selected_features_names = indep_X.columns[selected_indices]
        return selectk_features,selected_features_names.to_list()
```

```python
def split_scalar(indep_X,dep_Y):
        X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size = 0.25, random_state = 0)
        sc = StandardScaler()
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)    
        return X_train, X_test, y_train, y_test,sc
```

```python
#Evaluation Metrics for Regression Models
def R2_prediction(regressor,x_test):
    y_pred = regressor.predict(x_test)
    
    from sklearn.metrics import r2_score
    R2_score = r2_score(y_test,y_pred)
    return regressor,R2_score,x_test,y_test
```

```python
def random(x_train,y_train,x_test):
    from sklearn.ensemble import RandomForestRegressor
    regressor = RandomForestRegressor(n_estimators = 10, criterion = 'mse', max_features = 'sqrt', random_state=0)
    regressor.fit(x_train,y_train)
    R2_score = R2_prediction(regressor,x_test)
    return regressor,R2_score,x_test,y_test
```

```python
def selectk_regression(r2rf):
    dataframe = pd.DataFrame(index=['ChiSquare'], columns = ['Random'])
    for number,index in enumerate(dataframe.index):
        dataframe['Random'][index] = r2rf[number] 
    return dataframe
```

```python
data=pd.read_csv("prep.csv")

data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data['classification_yes'].value_counts()
indep_X=data.drop('classification_yes',axis=1)
dep_Y=data['classification_yes']
```

```python
kbest,selected_features=selectkbest(indep_X,dep_Y,6)
```

```python
x_train,x_test,y_train,y_test,sc = split_scalar(kbest,dep_Y)
```

```python
r2rf=[]
```

```python
regressor,R2_score,x_test,y_test = random(x_train,y_train,x_test)
r2rf.append(R2_score[1])

result = selectk_regression(r2rf)
```
